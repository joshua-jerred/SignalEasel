```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

for filename in ["in1.csv", "in2.csv"]:
    in2 = pd.read_csv(filename)
    
    plt.figure().set_figwidth(15)
    
    
    combo = (in2['s1']+in2['s2']) 
    combo = moving_average(combo, 100)
    combo = combo/2
    
    variance = combo.var()
    
    print(combo.mean() / (2**16), variance / (2**16))
    
    plt.plot(in2['s1'])
    plt.plot(in2['s2'])
    plt.plot(combo, color='red')
    plt.title("Skewed Data")
    plt.show()
```
