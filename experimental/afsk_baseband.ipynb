```python
import matplotlib.pyplot as plt
import numpy as np

SAMPLE_RATE = 48000
BAUD_RATE = 1200
SAMPLES_PER_BAUD = SAMPLE_RATE / BAUD_RATE
print(SAMPLES_PER_BAUD)

data = np.loadtxt('out.txt')
data = data[0:1000]
crossing = []

last = data[0]
counter = 0
crossing_distance = []
for point in data:
    if (last < 0 and point > 0) or (last > 0 and point < 0): # Crossing
        crossing.append(1)
        crossing_distance.append(counter)
        counter = 0
    else:
        crossing.append(0)
    last = point
    counter += 1

print(crossing_distance)
avg = 0
for num in crossing_distance:
    avg += num
avg /= len(crossing_distance) - 1
print(avg)
```

```python
plt.plot(data)
plt.plot(crossing)
plt.xlabel('x')
plt.ylabel('y')
plt.figure().set_figwidth(15)
plt.show()
```
